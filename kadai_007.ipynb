```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sample_pandas_6.csv")
df['発注日'] = pd.to_datetime(df['発注日'], errors='coerce')
df['month'] = df['発注日'].dt.month
df['season'] = df['month'] % 12 // 3 + 1
df['sales'] = df['単価'] * df['注文数']

season_sales = df.groupby('season')['sales'].sum().reindex([1,2,3,4])

x = [1,2,3,4]
plt.figure(figsize=(6,4))
plt.plot(x, season_sales.values, marker='o', linewidth=2, color='tab:blue')
plt.xticks(x, ['Winter','Spring','Summer','Autumn'])
plt.xlabel('Season')
plt.ylabel('Sales (JPY)')
plt.title('Seasonal Sales Trend')
plt.ylim(0)  # y-axis starts from 0
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib  # 日本語対応

# データ読み込み
df_sales = pd.read_csv("sample_pandas_6.csv")
df_cat = pd.read_csv("category.csv")

# 商品番号で結合
df = pd.merge(df_sales, df_cat, on='商品番号', how='left')

# 1. カテゴリーごとの出現頻度
category_counts = df['カテゴリー'].value_counts()

plt.figure(figsize=(6,4))
category_counts.plot(kind='bar', color='tab:green')
plt.xlabel('カテゴリー')
plt.ylabel('件数')
plt.title('カテゴリーごとの商品数')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 商品番号ごとの注文数基本統計量
stats_by_item = df.groupby('商品番号')['注文数'].describe()
print(stats_by_item)
```
