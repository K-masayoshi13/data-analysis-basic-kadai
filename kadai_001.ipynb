```python
from openpyxl import Workbook
from datetime import datetime

# ファイル名（現在日付）
today = datetime.now().strftime("%Y%m%d")
date_str = datetime.now().strftime("%Y/%m/%d")
fname = f"請求書_{today}.xlsx"

wb = Workbook()
ws = wb.active

ws["B2"] = "請求書"
ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名:鈴木一郎 様"

ws["F4"] = "No."
ws["G4"] = "0001"
ws["F5"] = "日付"
ws["G5"] = date_str

ws["B10"] = "商品名"
ws["C10"] = "数量"
ws["D10"] = "単価"
ws["E10"] = "金額"

ws.append([None, "商品A", 2, 10000, 20000])
ws.append([None, "商品B", 1, 15000, 15000])

ws["E13"] = 35000
ws["B15"] = "小計"
ws["E15"] = 35000
ws["B16"] = "消費税"
ws["E16"] = 3500
ws["B17"] = "合計"
ws["E17"] = 38500

wb.save(fname)
```
