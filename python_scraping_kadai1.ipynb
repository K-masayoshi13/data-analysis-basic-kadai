```python
import requests
from bs4 import BeautifulSoup

url = "https://news.yahoo.co.jp/articles/b8ab65fadb7ea9a625782ff4468e38ebfa7aae4f"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

body = soup.select_one("#uamods > div:nth-of-type(1) > div:nth-of-type(1) > p")

text = body.get_text(strip=True)

print(text)
```
