```python
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1, 2, 3]
y = [1, 2, -5, 2]
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)

# 
np.random.seed(3)
x2 = 4 + np.random.normal(0, 2, 60)
y2 = 4 + np.random.normal(0, 2, len(x2))

fig, axs = plt.subplots(2, 3, tight_layout=True)

# 
axs[0, 0].plot(x,y)

# sin
axs[0, 1].plot(x1, y1)

# 
axs[0, 2].hist(np.random.randn(100), bins=30)

# 
axs[1, 0].scatter(x2,y2)


plt.show()
```
